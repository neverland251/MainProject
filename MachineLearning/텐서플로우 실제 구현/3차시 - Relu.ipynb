```python
import tensorflow as  tf
import numpy as np

init = tf.global_variables_initializer()
```

```python
##relu는 투입값이 0보다 크면 그 출력값을 그대로 반환하고, 

n_features = 3
```

```python
#변수 주입기. eval시에 feed_dict 옵션을 사용해 데이터를 주입한다.
# shape의 None은 해당 데이터의 제한을 두지 않는다는 의미이다. 여기서는, 행은 상관하지 않겠다는 의미이다.

X = tf.placeholder(tf.float32,shape=(None,n_features),name="x")
```

```python
#가중치를 랜덤으로 초기화한다.

w1 = tf.Variable(tf.random_normal([n_features,1]),name="weights1")
w2 = tf.Variable(tf.random_normal([n_features,1]),name="weights2")
```

```python
#편향은 0을 부여한다.

b1 = tf.Variable(0.0,name="bias1")
b2 = tf.Variable(0.0,name="bias2")
```

```python
# WTX+B꼴의 선형 방정식을 만들어준다.

z1 = tf.add(tf.matmul(X,w1),b1,name="z1")
z2 = tf.add(tf.matmul(X,w2),b1,name="z2")
```

```python
#r(WTX+B)로 해당 선형방정식을 함수화한다. maximum 함수는 해당 값과 0 사이에 큰값을 선택하도록 하는 옵션이다.

relu1 = tf.maximum(z1,0.0,name="relu1")
relu2 = tf.maximum(z2,0.0,name="relu2")
```

```python
#relu1과 relu2 산출물의 결과를 더해서 출력한다.

output = tf.add(relu1,relu2,name="output")
```

```python
with tf.Session() as sess:

    sess.run(init)
    returns = output.eval(feed_dict = {X:[[2,5,8]]})
```

```python
### 쓸데없이 반복하지 않도록, 이를 함수화하자.

def relu(X):
    print(X.get_shape())
    n_shape = (int(X.get_shape()[1]),1)
    w = tf.Variable(tf.random_normal(n_shape),name = "weights")
    b = tf.Variable(0.0,name="bias")
    z = tf.add(tf.matmul(X,w),b,name="z")
    return tf.maximum(z,0.0,name="relu")
    
```

```python
init = tf.global_variables_initializer()

X = tf.placeholder(tf.float32,shape=[None,n_features],name = "X")
relus = [relu(X) for i in range(5)]
output = tf.add_n(relus,name="output")
```

```python
with tf.Session() as sess:
    sess.run(init)
    sess.run(output, feed_dict={X:[[1,2,3]]})
```
